## Objetivos del trabajo:  

Nustro objetivo es desarollar un modelo de machine learning que pueda predecir el comportamiento de las empresas del S&P500.
Este modelo predictivo permitiría: 

Anticipar movimientos del mercado.

Identificar empresas con mayor probabilidad de superar el rendimiento promedio del índice.

Optimizar decisiones de inversión.

Intentar hacernos ricos 

A pesar de que en la asignatura de tecnicas de predicción hemos dados modelos econometricos como el ARIMA. El machine learning nos permitiria capturar las relaciones no linieales que hay en el mercado, buscando correlaciones y datos del mercado que no se detecten con modelos tradicionales.

## Análisi del dataset:

El datase ha sido obtenido de Kaggle:

https://www.kaggle.com/datasets/codebynadiia/s-and-p-500-stocks-dataset-first-half-2025

El dataset contiene información diaria del S&P500 entre enero y junio de 2025, e incluye las siguientes variables principales:
 Variables de identificación

company_name: nombre de la empresa.

ticker: símbolo bursátil.

 Variables temporales

(El dataset presenta columnas con formato dd-mm-yyyy_tipo)

_opening → precio de apertura del día

_closing → precio de cierre del día

_volume → volumen transado (número de acciones negociadas)

No presenta valores nulos en las series de precios.

No existen precios negativos o inconsistentes.

El volumen también está completo para todas las empresas analizadas.

Las fechas están correctamente estructuradas, aunque el dataset viene en formato WIDE.

Creeemos que el dataset tiene la calidad suficiente para intentar realizar técnicas de machine learning.
