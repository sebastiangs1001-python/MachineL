```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Filtrar aos
df20 = Ipc_u[Ipc_u['Year']==2020].sort_values('Country')
df24 = Ipc_u[Ipc_u['Year']==2024].sort_values('Country')

# Asegurar el mismo orden de pases
countries = df20['Country'].values
x = np.arange(len(countries))

y20 = df20['GDP_USD_Billions'].values
y24 = df24['GDP_USD_Billions'].values
```

```python
plt.figure(figsize=(14,6))

plt.bar(x, y20, color='blue', alpha=0.5, edgecolor='black', linewidth=1.3, label='2020')
plt.bar(x, y24, color='red', alpha=0.5, edgecolor='black', linewidth=1.3, label='2024')

plt.xticks(x, countries, rotation=90)
plt.title("PIB 2020 vs 2024 (Barras superpuestas con borde)")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(14,6))

plt.bar(x, y20, color='blue', alpha=0.4, label='2020')
plt.bar(x, y24, color='red', alpha=0.4, label='2024')

# Diferencias
diff = y24 - y20

for i, d in enumerate(diff):
    arrow = '' if d > 0 else ''
    color = 'green' if d > 0 else 'red'
    plt.text(x[i], max(y20[i], y24[i]) + 0.05*max(y24), 
             f"{arrow} {abs(round(d,2))}", 
             ha='center', color=color, fontsize=9)

plt.xticks(x, countries, rotation=90)
plt.title("PIB 2020 vs 2024 (Con flechas de crecimiento/cada)")
plt.legend()
plt.show()
```

```python
pct_change = (y24 - y20) / y20 * 100

plt.figure(figsize=(14,6))
plt.bar(x, pct_change, color='purple', alpha=0.7)

plt.axhline(0, color='black', linewidth=1)
plt.xticks(x, countries, rotation=90)
plt.ylabel("% cambio (2020  2024)")
plt.title("Cambio porcentual del PIB entre 2020 y 2024 (%)")
plt.show()
```

```python
year = 2024
df_year = Ipc_u[Ipc_u['Year'] == year]

unemployment = df_year['Unemployment_Rate_%'].mean()
employment = 100 - unemployment

plt.figure(figsize=(6,6))
plt.pie(
    [unemployment, employment],
    labels=['Desempleados', 'Empleados'],
    autopct='%1.1f%%',
    colors=['red', 'green'],
    startangle=90
)
plt.title(f"Empleados vs Desempleados (Promedio global)  {year}")
plt.show()
```

```python
year = 2020
df_year = Ipc_u[Ipc_u['Year'] == year]

countries = df_year['Country'].unique()
n = len(countries)

fig, axes = plt.subplots(
    nrows=(n//3)+1, ncols=3, 
    figsize=(14, (n//3)*4)
)

axes = axes.flatten()

for i, country in enumerate(countries):
    df_c = df_year[df_year['Country'] == country]
    unemp = df_c['Unemployment_Rate_%'].values[0]
    emp = 100 - unemp

    axes[i].pie(
        [unemp, emp],
        labels=['Desempleados', 'Empleados'],
        autopct='%1.1f%%',
        colors=['red', 'green'],
        startangle=90
    )
    axes[i].set_title(country)

# Apagar ejes vacos
for j in range(i+1, len(axes)):
    axes[j].axis('off')

plt.tight_layout()
plt.show()
```
