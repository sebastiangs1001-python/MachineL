```python
# 1. Convertir de wide a long (melt)
sp5L = sp500.melt(
    id_vars=["company_name", "ticker"],
    var_name="feature",
    value_name="value"
)

# 2. Separar la fecha y la métrica
sp5L[["date", "metric"]] = sp5L["feature"].str.rsplit("_", n=1, expand=True)

# 3. Convertir la fecha a tipo datetime
sp5L["date"] = pd.to_datetime(sp5L["date"], format="%d-%m-%Y")

# 4. Reordenar columnas (opcional, pero profesional)
sp5L = sp5L[["company_name", "ticker", "date", "metric", "value"]]

sp5L.head(7)
```
