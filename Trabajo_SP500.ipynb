```python

top10 = sp500.head(10 )["company_name"].tolist()
close_cols = sp500.filter(like="closing")
plt.figure(figsize=(16,8))
final_values = {}


for company in top10:
    
    idx = sp500[sp500['company_name'] == company].index[0]
    series = close_cols.iloc[idx]

    dates = series.index.str.replace("_closing", "")
    dates = pd.to_datetime(dates, format="%d-%m-%Y")

    # Normalización
    normalized = (series.values / series.values[0]) * 100
    final_values[company] = normalized[-1]

    plt.plot(dates, normalized, label=company, linewidth=1.8)
    plt.text(
        dates[-1],
        normalized[-1],
        f"{normalized[-1]:.1f}",
        fontsize=9,
        va='center'
    )


sp500_series = close_cols.mean(axis=0)
sp500_dates = sp500_series.index.str.replace("_closing", "")
sp500_dates = pd.to_datetime(sp500_dates, format="%d-%m-%Y")

sp500_norm = (sp500_series.values / sp500_series.values[0]) * 100
final_values["S&P 500 Benchmark"] = sp500_norm[-1]

plt.plot(sp500_dates, sp500_norm, color="black", linewidth=3, label="S&P 500 Benchmark")

plt.text(
    sp500_dates[-1],
    sp500_norm[-1],
    f"{sp500_norm[-1]:.1f}",
    fontsize=10,
    va='center',
    fontweight='bold',
    color="black"
)

plt.title("Evolución Normalizada del Precio — Top 10 Empresas + S&P 500 (Base = 100)")
plt.xlabel("Fecha")
plt.ylabel("Índice Normalizado (Inicio = 100)")
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

print("\nCrecimiento Final desde 100:\n")
for k, v in final_values.items():
    print(f"{k:25s}  {v:.2f}")
```

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

company = "Nvidia"

# 1. Tomar columnas de cierre
close_cols = sp500.filter(like="closing")

# 2. Fila de la empresa
idx = sp500[sp500["company_name"] == company].index[0]

# 3. Convertir a serie temporal
series = close_cols.iloc[idx]
dates = pd.to_datetime(series.index.str.replace("_closing",""), format="%d-%m-%Y")
ts = pd.Series(series.values, index=dates)

# 4. Graficar ACF y PACF
fig, ax = plt.subplots(1,2, figsize=(14,4))

plot_acf(ts, ax=ax[0])
ax[0].set_title(f"ACF — {company}")

plot_pacf(ts, ax=ax[1])
ax[1].set_title(f"PACF — {company}")

plt.tight_layout()
plt.show()
```

```python
# 1. Seleccionar columnas de cierre
close_cols = sp500.filter(like="closing")

# 2. Poner empresas como columnas y fechas como filas
close_matrix = close_cols.T
close_matrix.columns = sp500["company_name"]

# 3. Matriz de correlaciones
corr = close_matrix.corr()

# 4. Heatmap
plt.figure(figsize=(10,8))
plt.imshow(corr, cmap="viridis", aspect="auto")
plt.colorbar()
plt.title("Correlograma de Empresas (Correlación de Precios)")
plt.xticks(range(len(corr)), corr.columns, rotation=90)
plt.yticks(range(len(corr)), corr.columns)
plt.tight_layout()
plt.show()
```

```python
emp1 = "Nvidia"
emp2 = "Microsoft"

# Columnas de cierre
close_cols = sp500.filter(like="closing")

# Fila de empresas
idx1 = sp500[sp500["company_name"] == emp1].index[0]
idx2 = sp500[sp500["company_name"] == emp2].index[0]

series1 = close_cols.iloc[idx1]
series2 = close_cols.iloc[idx2]

# Convertir fechas
dates = series1.index.str.replace("_closing", "")
dates = pd.to_datetime(dates, format="%d-%m-%Y")

# Series temporales
ts1 = pd.Series(series1.values, index=dates)
ts2 = pd.Series(series2.values, index=dates)

# Calcular retornos
ret1 = ts1.pct_change()
ret2 = ts2.pct_change()

# Rolling correlation de 30 días
roll_corr = ret1.rolling(30).corr(ret2)

plt.figure(figsize=(12,5))
plt.plot(roll_corr.index, roll_corr.values)
plt.title(f"Rolling Correlation (30 días) — {emp1} vs {emp2}")
plt.xlabel("Fecha")
plt.ylabel("Correlación")
plt.grid(True)
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import matplotlib.pyplot as plt

# 1. Seleccionar top 10 empresas
top10 = sp500["company_name"].head(10).tolist()

# 2. Seleccionar precios de cierre
close_cols = sp500.filter(like="closing")  # CORREGIDO

# 3. Matriz de precios
close_matrix = []

for comp in top10:
    idx = sp500[sp500["company_name"] == comp].index[0]
    close_matrix.append(close_cols.iloc[idx].values)

close_matrix = np.array(close_matrix)

# 4. Matriz de correlaciones
corr = np.corrcoef(close_matrix)

# 5. Malla para el gráfico 3D
X, Y = np.meshgrid(range(len(top10)), range(len(top10)))
Z = corr

# 6. Gráfico 3D
fig = plt.figure(figsize=(12,8))
ax = fig.add_subplot(111, projection="3d")

ax.plot_surface(X, Y, Z, cmap="viridis", edgecolor="none")  # puedes usar plasma/coolwarm
ax.set_title("Correlograma 3D — Top 10 empresas")

ax.set_xticks(range(len(top10)))
ax.set_xticklabels(top10, rotation=45)

ax.set_yticks(range(len(top10)))
ax.set_yticklabels(top10, rotation=45)

ax.set_zlabel("Correlación")

plt.show()
```
